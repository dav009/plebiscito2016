```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
municipios_y_victimas = pickle.load(open('./plebicito_municipios_y_victimas20161009.pickle', 'rb'))
municipios_y_victimas['Si']= municipios_y_victimas['Si'].apply(int)
municipios_y_victimas['No']= municipios_y_victimas['No'].apply(int)
municipios_y_victimas['TOTAL']=municipios_y_victimas['TOTAL'].apply(int)
municipios_y_victimas['habilitados']=municipios_y_victimas['habilitados'].apply(int)
```

```python
municipios_y_victimas['CocienteSi'] = municipios_y_victimas['Si']/municipios_y_victimas['habilitados'] 
municipios_y_victimas['CocienteNo'] = municipios_y_victimas['No']/municipios_y_victimas['habilitados'] 
municipios_y_victimas['CocienteVictimas'] = municipios_y_victimas['TOTAL']/municipios_y_victimas['habilitados']
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(municipios_y_victimas['CocienteVictimas'], municipios_y_victimas['CocienteSi'], 's', label='votos Si')
ax.plot(municipios_y_victimas['CocienteVictimas'], municipios_y_victimas['CocienteNo'], 'x', label='votos No')
ax.set_xlabel('# de victimas / # de personas habilitadas para votar')
ax.set_ylabel('votos (no/si)/ # de personas habilitadas para votar')
ax.legend()
```

```python
fig, ax = plt.subplots(1,1)
ax.semilogx(municipios_y_victimas['CocienteVictimas'], municipios_y_victimas['CocienteSi'], 's', label='votos Si')
ax.semilogx(municipios_y_victimas['CocienteVictimas'], municipios_y_victimas['CocienteNo'], 'x', label='votos No')
ax.set_xlabel('# de victimas / # de personas habilitadas para votar')
ax.set_ylabel('votos (no/si)/ # de personas habilitadas para votar')
ax.legend()
```
